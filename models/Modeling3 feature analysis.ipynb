```python
X = pd.read_csv('../data/processed/feature_matrix_scaled_PT.csv')
X.drop(columns='Unnamed: 0', inplace=True)
```

```python
X.head()
```

```python
rf_feats = feats.Features
```

```python
X[rf_feats]
```

```python
y = pd.read_csv('../data/processed/target_array_2.csv', index_col=0)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X[rf_feats], y.values.ravel(), test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
import time

start = time.time()

rf = RandomForestClassifier(n_jobs=-1)
rf.fit(X_train[:100000], y_train[:100000])

end = time.time()
print('Time:', (end-start)/60, 'minutes')
```

```python
from sklearn.metrics import f1_score
ypred_rf = rf.predict(X_test)
f1_score(y_test, ypred_rf, average='weighted')
```

    10,000 samples: 0.5829661407004904

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(rf, X_test, y_test, labels=['very low', 'low','medium','high','very high'])
```

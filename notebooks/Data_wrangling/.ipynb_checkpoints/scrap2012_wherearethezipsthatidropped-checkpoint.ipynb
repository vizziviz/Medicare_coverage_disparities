##### Updating "Number_race_unknown" to include anyone not counted in other race bins

```python
def update_race_unknown(df, race_columns, race_unknown_column, total_column):
    sums = df[race_columns].sum(axis=1)
    idx = df[sums != df[total_column]].index
    
    known_race_sums = sums - df[race_unknown_column]
    new_values = df.loc[idx,total_column] - known_race_sums
    
    df.loc[idx, race_unknown_column] = new_values
```

##### Zip Codes

How many unique 5-digit zip codes are there in this df? There are 33,000 zip codes on Census df.

First I have to convert the NPPES Zip Code to a 9-digit code (with leading zeros) or else I'll incorrectly capture the five digit zip code.

```python
med.full_zip.apply(lambda x: len(str(x))).value_counts()
```

```python
def make_five_digit_zip(df, zipcol):
    #adding leading zeros if zip if less than 5 digits
    idx = df[df[zipcol].apply(lambda x: len(str(x)) < 5)].index
    df.loc[idx,zipcol] = df.loc[idx, zipcol].apply(lambda x: str(x).zfill(5))
    
    #adding leading zeros if zip is between 5 and 9 digits
    idx = df[df[zipcol].apply(lambda x: 5 < len(str(x)) < 9)].index
    df.loc[idx, zipcol] = df.loc[idx, zipcol].apply(lambda x: str(x).zfill(9))
    
    #make 9-digit zip into 5-digit zip
    five_digit_series = df[zipcol].astype('str').apply(lambda x: x[:5])
    
    return five_digit_series
```

```python
med['five_digit_zip'] = make_five_digit_zip(med, 'full_zip')
```
